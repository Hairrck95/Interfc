```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def Logistico(r,x):
    return r*x*(1-x)
```

```python
r = 0.5
# x = np.arange(0,1,0.01)
x = np.linspace(0,1,100)

y = Logistico(r,x)
```

```python
# plt.plot(x,y,'r.')
plt.plot(x,y)
```

```python
r = 0.5
x0 = 0.4  # x0

x1 = Logistico(r,x0)
print(x1)

x2 = Logistico(r,x1)
print(x2)

x3 = Logistico(r,x2)
print(x3)
```

## $x = rx(1-x)$


## $x = rx-rx^2$

## $x_n = rx_{n-1}(1-x_{n-1})$

```python
r = 1
x = 0.9  # x0
N = 5


for n in range(N):
    x = Logistico(r,x)
    print(x)
```

```python
# r = 3.84444
r = 3.5


x = 0.4  # x0
N = 200

ene = []
pob = []

for n in range(N):
    x = Logistico(r,x)
    # print(x)
    pob.append(x)
    ene.append(n)
plt.figure(figsize=(12,4))
plt.plot(ene,pob,'r.')
plt.plot(ene,pob, alpha=0.5)
plt.axhline(y = x, xmin = 0.0, color = 'blue', alpha=0.5)
plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')

plt.grid()
```
