```python
def f1(x):
    return x**4 - 2*x + 1

def I1(x):
    return (x**5)/5 - x**2 + x

def f3(x):
    return np.sin(x)

def I3(x):
    return -np.cos(x)


def f2(x):
    return np.exp(x**2) + 1/(x**2) + np.log(x)
```

```python
def IntTrapecio(f,a,b,N):

    s = 0.0
    h = (b-a)/N

    s = s + (f(a)+f(b))/2

    for k in range(1,N):
        s = s + f(a+k*h)

    s = h*s

    return float(s)




def IntSimpson(f,a,b,N):

    s = 0.0
    h = (b-a)/N

    s = s + (f(a)+f(b))

    for k in np.arange(1,N/2+1):
        s = s + 4*f(a+(2*(k-1))*h)

    for k in np.arange(1,N/2):
        s = s + 2*f(a+(2*k*h))
    
    s = (h/3)*s

    return float(s)
    
```

```python
a = 0
b = np.pi
N = 1000000

print(IntTrapecio(f3,a,b,N))
print(IntSimpson(f3,a,b,N))

```
