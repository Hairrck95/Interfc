```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
```

```python
def DifsAdelante(h,f,x0):
    return (f(x0+h)-f(x0))/h

def DifsAtras(h,f,x0):
    return (f(x0)-f(x0-h))/h

def DifsCentral(h,f,x0):
    return (f(x0+h/2)-f(x0-h/2))/h
```

```python
def DifCent2a(h,f,x0):
    return (f(x0+h)-2*f(x0)+f(x0-h))/h**2
```

```python
def f1(x):
    return 9*x**3 + np.cos(x)*x**5 - 2*x**2 + 5*x + 5000

def f2(x):
    return np.sin(x)

def f3(x):
    return x**2

def f4(x):
    return x*np.sin(x)
```

```python
h = 0.01
X = np.linspace(-10,10,100)

Fx = f4(X)
dX = DifsCentral(h,f4,X)
dX2 = DifsAdelante(h,f4,X)
dX3 = DifsAtras(h,f4,X)
```

```python

# plt.plot(X,Fx,'m.')
# plt.plot(X,dX,'r.')
plt.plot(X,Fx,'m', label = "F(x)")
plt.plot(X,dX,'r', label = "difs Cent")
plt.plot(X,dX2,'g')
plt.plot(X,dX3,'b')
plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.grid()
plt.legend()
```
